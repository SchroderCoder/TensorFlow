```python
# Cargar el archivo CSV
# Preprocesar los datos
df = pd.read_csv('quora.csv')
def clean_text(text):
    if isinstance(text, str):
        text = text.lower()
        text = re.sub(r'[?().]', '', text)
    return text

df = df[df['question1'].notna()]
question1 = df['question1'].apply(clean_text).tolist()
question2 = df['question2'].apply(clean_text).tolist()
```

```python

all_bows = []
for row1, row2 in zip(question1, question2):
    if isinstance(row1, str) and isinstance(row2, str):
        set1 = set(row1.lower().split())
        set2 = set(row2.lower().split())
        bow = set1.union(set2)
        all_bows.append(bow)
    else:
        all_bows.append(set())

print(all_bows)
```

```python
results = []

for q1, q2, bow in zip(question1, question2, all_bows):
    words_q1 = q1.split()
    words_q2 = q2.split()

    sorted_bow = sorted(bow)
    vector_q1 = [words_q1.count(word) for word in sorted_bow]
    vector_q2 = [words_q2.count(word) for word in sorted_bow]

    vector_q1 = vector_q1/np.linalg.norm(vector_q1)
    vector_q2 = vector_q2/np.linalg.norm(vector_q2)

    results.append(np.dot(vector_q1, vector_q2))

results_df = pd.DataFrame(results, columns=["similarity"])

results_df.head(1000).to_csv('resultados_bow.csv')

```
